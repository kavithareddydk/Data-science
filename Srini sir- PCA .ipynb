```python
univ= pd.read_csv("Universities.csv")
```

```python
univ.head()
```

```python
## considering only numerical data
univ1 = univ.iloc[:,1:]
univ1
```

```python
univ1.info()
```

```python
## standardisation of the numerical data
std = StandardScaler()
univ1_normal= std.fit_transform(univ1)
univ1_normal
```

```python
## Norm = MinMaxScaler()
## univ1_normal = norm.fit_transform(univ1)
# univ1_normal
```

```python
pca2 =PCA(n_components=None) ## 
data =pca2.fit_transform(univ1_normal)
pca_components = pd.DataFrame(data, columns = ['PC1', 'PC2','PC3','PC4','PC5','PC6'])
pca_components
```

```python
## note ; In the above values, PC1 CONTAINNS MAXIMUM INFORMATION WHICH WE CAN CAN CONSIDER FOR FUTURE CALCULATIONS 
# and on PC components then we get correlation =0 . thts true 
# scatter plot - no trend wil be seen among PCA components
```

```python
pca_components.corr()
## DIAGONAL MATRIX SHOWS 1 CORRELATION
# HERE ARLL CORRELATION VALUES ARE ALMOST 0....VERY LESS VALUE
# NOTE: ON PRINCIPAL COMPONENTS , THE CORRELATION WILL BE ZERO
```

```python
plt.scatter(pca_components.PC1, pca_components.PC2)
# HERE NO TREND IS THERE
```

```python
#IF WE WANT ONLY 3PCA COMPONENTS , we shoulg give here3
pca3 =PCA(n_components=3) ## 
data =pca3.fit_transform(univ1_normal)
data
```

```python
## the amount of variance that each PCA explains is given by the EIGEN VALUES
var = pca3.explained_variance_ratio_
var
```

```python
pca2 =PCA(n_components=None) ##  it wil give all 6 components values
data =pca2.fit_transform(univ1_normal)
data
```

```python
## the amount of variance that each PCA explains is given by the EIGEN VALUES
var = pca2.explained_variance_ratio_
var
```

```python

## CUMULATIVE VARIANCE 
var1 = np.cumsum(np.round(var, decimals=4)*100)
var1
```

```python
#     note : np.cumsum([10,15,25,35])
# it is cumulative sum = 10,25,50,85fl
```

```python
pca2.components_ ## these are all eigen vectors ,( for information given here)
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
plt.title('Scree plot')
plt.xlabel('PC')
plt.ylabel('Cumulative Variance Ratio')
plt.grid()
plt.show()
```

```python
## SCREE PLOT: ITS A CUMULATIVE VARIANCE V/S PC PLOT 
```
